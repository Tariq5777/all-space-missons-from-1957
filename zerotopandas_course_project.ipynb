## Data Preparation and Cleaning

As I have downloaded dataset to the notebook, now we will prepare and clean up the dataset for the exploratory ananlyisis of the dataset. 
Data preparation is the process of cleaning and transforming raw data prior to processing and analysis. It is an important step prior to processing and often involves reformatting data, making corrections to data and the combining of data sets to enrich data.

```python
import pandas as pd
```

```python
space_missions_df = pd.read_csv('./all-space-missions-from-1957/Space_Corrected.csv')
```

```python
space_missions_df
```

```python
space_missions_df.info()
```

```python
import numpy as np
```

```python
print(space_missions_df["Company Name"].nunique())
```

There are 56 diffrent companies which have done space missions since 1957.

```python
space_missions_df = space_missions_df.rename(columns = {" Rocket":"Rocket_Cost"})
space_missions_df = space_missions_df.rename(columns = {"Company Name":"Company_Name"})
space_missions_df = space_missions_df.rename(columns = {"Status Rocket":"Status_Rocket"})
space_missions_df = space_missions_df.rename(columns = {"Status Mission":"Status_Mission"})
```

```python
space_missions_df['Day']=space_missions_df['Datum'].apply(lambda datum: datum.split()[0])
space_missions_df['Month']=space_missions_df['Datum'].apply(lambda datum: datum.split()[1])
space_missions_df['Date']=space_missions_df['Datum'].apply(lambda datum: datum.split()[2][:2]).astype(int)
space_missions_df['Year']=space_missions_df['Datum'].apply(lambda datum: datum.split()[3]).astype(int)
space_missions_df['Hour']=space_missions_df['Datum'].apply(lambda datum: int(datum.split()[-2][:2]) if datum.split()[-1]=='UTC' else np.nan)
space_missions_df['Country'] = space_missions_df['Location'].apply(lambda location: location.split(',')[-1])
space_missions_df.drop(['Datum', 'Detail', 'Unnamed: 0', 'Unnamed: 0.1'], 1, inplace=True)
```

```python
space_missions_df["Rocket_Cost"] = space_missions_df["Rocket_Cost"].str.replace(',', '');
```

```python
space_missions_df["Rocket_Cost"]= space_missions_df["Rocket_Cost"].astype(float)
```

## Asking and Answering Questions

#### We have seen some visualizations on the basis of the columns, we will now answer the questions based on our dataset, which will help us to know about our dataset.

#### Q1: Show the data of the launches made in 2020 with respect to the countries.

```python
plt.figure(figsize=(20,10))
df_latest=space_missions_df[space_missions_df['Year']==2020]
sns.catplot(x ='Country',
            data=df_latest,
            kind='count',
            aspect=2,
            height=8)
plt.title("Country Space Mission Count 2020", fontsize = 20)
plt.xticks(size=10)
plt.xlabel('Country',fontsize = 15)
plt.ylabel('Count',fontsize = 15)
plt.yticks(size=10)
```

### sns.catplot()
This function provides access to several axes-level functions that show the relationship between a numerical and one or more categorical variables using one of several visual representations. This function always treats one of the variables as categorical and draws data at ordinal positions (0, 1, … n) on the relevant axis, even when the data has a numeric or date type.

#### Q2: Show all the status missions for top 5 companies ?

```python
successMissions=pd.crosstab(space_missions_df["Company_Name"],space_missions_df["Status_Mission"]).sort_values(by='Success',ascending=False).reset_index().head(5)

failedMissions = pd.crosstab(space_missions_df["Company_Name"],space_missions_df["Status_Mission"]).sort_values(by='Failure',ascending=False).reset_index().head(5)

partialFailedMissions=pd.crosstab(space_missions_df["Company_Name"],space_missions_df["Status_Mission"]).sort_values(by='Partial Failure',ascending=False).reset_index().head(5)

prelaunchFailedMissions=pd.crosstab(space_missions_df["Company_Name"],space_missions_df["Status_Mission"]).sort_values(by='Prelaunch Failure',ascending=False).reset_index().head(5)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15,15))


axes[0,0].set_title('Company vs Success Mission',fontsize = 15)
sns.barplot(
    data=successMissions, 
    y="Success", x="Company_Name", hue="Success",
    ci="sd", palette="dark", alpha=.6, ax = axes[0,0]
)

axes[0,1].set_title('Company vs Failed Mission',fontsize = 15)


sns.barplot(
    data=failedMissions, 
    y="Failure", x="Company_Name", hue="Failure",
    ci="sd", palette="dark", alpha=.6,  ax=axes[0,1]
)

axes[1,0].set_title('Company vs Partial Failed Mission', fontsize = 15)



sns.barplot(
    data=partialFailedMissions, 
    y="Partial Failure", x="Company_Name", hue="Partial Failure",
    ci="sd", palette="dark", alpha=.6,  ax =axes[1,0]
)

axes[1,1].set_title('Company vs Prelaunch Failed Mission',fontsize = 15)


sns.barplot(
    data=prelaunchFailedMissions, 
    y="Prelaunch Failure", x="Company_Name", hue="Prelaunch Failure",
    ci="sd", palette="dark", alpha=.6, ax = axes[1,1]
)
```

### sns.barplot()

A bar plot represents an estimate of central tendency for a numeric variable with the height of each rectangle and provides some indication of the uncertainty around that estimate using error bars. Bar plots include 0 in the quantitative axis range, and they are a good choice when 0 is a meaningful value for the quantitative variable, and you want to make comparisons against it.

For datasets where 0 is not a meaningful value, a point plot will allow you to focus on differences between levels of one or more categorical variables.

It is also important to keep in mind that a bar plot shows only the mean (or other estimator) value, but in many cases it may be more informative to show the distribution of values at each level of the categorical variables. In that case, other approaches such as a box or violin plot may be more appropriate.

#### We can see here that RVSN USSR is leading in all the plots.

#### Q3: Compare the launches done by NASA and ISRO .

```python
plt.figure(figsize=(12,5))
df1 = pd.DataFrame({'count_company_1' : space_missions_df[space_missions_df['Company_Name']=='ISRO'].groupby(by='Year').size()})
df2 = pd.DataFrame({'count_company_2' : space_missions_df[space_missions_df['Company_Name']=='NASA'].groupby(by='Year').size()})
plt.plot(df1['count_company_1'], 'o-r')
plt.plot(df2['count_company_2'], 'o-b')
plt.title('NASA vs ISRO (Launches)',fontsize = 15)
plt.legend(("ISRO", "NASA"))
```

### plt.plot()

#### We can see here that NASA started doing space missions before 1960 and ISRO started its space mission just before 1980, NASA has done more space missions than ISRO, but in the recent years ISRO is taking up the pace.

#### Q4: Show the plot between money spent with respect to the companies.

```python
highestCost = space_missions_df.copy()
df_cost=highestCost.dropna()
df_cost_comp=df_cost.groupby('Company_Name')['Rocket_Cost'].sum().reset_index().sort_values(by='Rocket_Cost',
                                                                                            ascending=False)
df_cost_comp['Rocket_Cost']=df_cost_comp['Rocket_Cost']/1000 #Converting costs to billion $
```

The copy function copies the whole dataframe to highestCost, then df_cost is assigned with all the values that are not null by using the dropna() function. Then we store all the values of the dataframe in df_cost to a new dataframe with Company Name and Rocket Cost summed up, and then we simply convert the float numbers into billion so it is easy to read for ther user.

```python
plt.figure(figsize=(9,9))
sns.catplot(x = 'Company_Name',
            y = 'Rocket_Cost',
            data=df_cost_comp,
            aspect=2,
            height=6,
            kind='point')
plt.xticks(size=12,rotation=45)
plt.xlabel('Company',size=20)
plt.ylabel('Money spent (in billion $)',size=20)
plt.yticks(size=15)
plt.title('Money spent on by Companies ',size=20)
```

#### We use the same catplot function we used earlier, but here we change the kind of the catplot from count to point.
#### We can clearly see that NASA is the company who have spent the highest amount of money compared to others.

#### Q5: Highest number of active rockets by companies, highest number of retired rockets by companies.

```python
active=space_missions_df[space_missions_df['Status_Rocket']=='StatusActive']
retired=space_missions_df[space_missions_df['Status_Rocket']=='StatusRetired']
```

```python
fig, axes = plt.subplots(2, 1, figsize=(16, 12))

axes[0].plot(active['Company_Name'].value_counts()[:15], 's-g')
axes[0].set_xlabel('Company')
axes[0].set_ylabel('No. of Active Rockets')
axes[0].set_title('Active Rockets by Companies', fontsize = 15)

axes[1].plot(retired['Company_Name'].value_counts()[:15],'s-r')
axes[1].set_xlabel('Company')
axes[1].set_ylabel('No. of Retired Rockets')
axes[1].set_title('Retired Rockets by Companies', fontsize = 15)
axes[1].tick_params(axis ='x', rotation =-45)
```

### plt.subplots()

Create a figure and a set of subplots.

This utility wrapper makes it convenient to create common layouts of subplots, including the enclosing figure object, in a single call.

We can create many plots in a single cell by using the subplots function, which is provided by the matplotlib  library.
